```python
import matplotlib.pyplot as plt

sequential_times = [32.6168, 18.4807, 14.7581, 18.9366, 33.054, 64.1952]
parallel_times = [30.9218, 7.78702, 2.11106, 0.864288, 0.274533, 0.164467]


N_values = [1, 2, 4, 8, 16, 32]

plt.figure(figsize=(10, 6))

plt.plot(N_values, sequential_times, marker='o', linestyle='-', color='blue', label='Sequencial')
plt.plot(N_values, parallel_times, marker='o', linestyle='-', color='red', label='Paralela')

plt.xscale('linear')
plt.yscale('linear')
plt.ylim(0, 40)

plt.xticks(N_values)

plt.xlabel('Número de Subvetores (N)')
plt.ylabel('Tempo de Execução (segundos)')
plt.title('Comparação entre Ordenação Sequencial e Paralela')
plt.legend()
plt.grid(True)
plt.show()
```
